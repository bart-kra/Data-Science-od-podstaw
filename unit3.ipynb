```python
from matplotlib import pyplot as  plt
from collections import Counter
```

```python
# SIMPLE CHART CREATING
years = [1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000,
          2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011,
          2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

pkb = [1731, 2236, 2459, 2497, 2875, 3687, 4147, 4123, 4518, 4398, 4501,
       4991, 5207, 5702, 6681, 8022, 9035, 11255, 13996, 11526, 12504,
       13777, 13011, 13558, 14182, 12560, 12379, 13816, 15505, 15700, 15817,
       18000, 18900]

plt.plot(years, pkb, color='green', marker='o', linestyle='solid')
plt.title('Poland PKB in 1990-2022')
plt.xlabel('year')
plt.ylabel('PKB ($)')
plt.xlim(years[0], years[-1])
plt.ylim(min(pkb), max(pkb))
plt.show()
```

```python
# CREATING SIMPLE BAR GRAPH
movies = ['Titanic', 'Forrest Gump', 'Die Hard', 'Ben-Hur', 'Dune', 'Hobbit']
oscars = [11, 6, 8, 11, 6, 1]

plt.bar(range(len(movies)), oscars)
plt.title('Best Awarded Movies')
plt.ylabel('Awards number')
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
grades = [4,2,4,5,33,45,76,34,67,87,65,32,76,32,12,46,1,79,97,43,76,8,54,99,76,98,65,46,21]
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()], histogram.values(), 10) # MIDDLING BARS USING +5
plt.axis([-5, 105, 0, 10]) # FITTING RANGES
plt.xticks([10 * i for i in range(11)]) # DECILES INDEXING
plt.xlabel('Decile')
plt.ylabel('Count')
plt.title('Grades histogram')
plt.show()
```
